```python
import pandas as pd
from utils.discretizer import discretize


def normalize(n, max_possible_point=7):
    return (1-(n/max_possible_point))*10

# Column: ["Year","Country","Number","First name","Last name","Middle name","Gender","P1","P2","P3","P4","Total score","Award"]
df = pd.read_csv('benelux/bxmo-results-all.csv')
df = df[["Year","P1","P2","P3","P4","Total score"]]
df = df.dropna()

agg_func = {
    'P1': 'mean',
    'P2': 'mean',
    'P3': 'mean',
    'P4': 'mean'
}
df = df.groupby('Year').agg(agg_func)
df = df.sort_values(by=['Year'], ascending=False)
df['P1'] = df['P1'].apply(normalize)
df['P2'] = df['P2'].apply(normalize)
df['P3'] = df['P3'].apply(normalize)
df['P4'] = df['P4'].apply(normalize)
df = df.reset_index()
score = df[['P1', 'P2', 'P3', 'P4']].to_numpy()
score = score.flatten()

display(df)
print(score)

#############################################
benelux = pd.read_csv('benelux/PROBLEMS_benelux.csv')
benelux['score'] = score
benelux['level'] = benelux['score'].apply(discretize)

benelux.to_csv('../data/regression/benelux/benelux_combined.csv')
benelux
```
