```python
import pandas as pd
import numpy as np


r = range(2023, 2012-1, -1)
df = [pd.read_csv(f"egmo/egmo{i}.csv") for i in r]
max_possible_point = 7
```

```python
# This cell is not used, ONLY for display
temp_dict = {
    "year": [i for i in r],
    "all": [],
    "normalized_p1": [],
    "normalized_p2": [],
    "normalized_p3": [],
    "normalized_p4": [],
    "normalized_p5": [],
    "normalized_p6": [],
    "normalized_p7": [],
    "normalized_p8": [],
}
for i, year in enumerate(temp_dict["year"]):
    if year == 2012:
        temp_dict["all"].append(df[i].shape[0])
        temp_dict["normalized_p1"].append((1-(df[i].P1.mean()/max_possible_point))*10)
        temp_dict["normalized_p2"].append((1-(df[i].P2.mean()/max_possible_point))*10)
        temp_dict["normalized_p3"].append((1-(df[i].P3.mean()/max_possible_point))*10)
        temp_dict["normalized_p4"].append((1-(df[i].P4.mean()/max_possible_point))*10)
        temp_dict["normalized_p5"].append((1-(df[i].P5.mean()/max_possible_point))*10)
        temp_dict["normalized_p6"].append((1-(df[i].P6.mean()/max_possible_point))*10)
        temp_dict["normalized_p7"].append((1-(df[i].P7.mean()/max_possible_point))*10)
        temp_dict["normalized_p8"].append((1-(df[i].P8.mean()/max_possible_point))*10)
        continue
    
    temp_dict["all"].append(df[i].shape[0])
    temp_dict["normalized_p1"].append((1-(df[i].P1.mean()/max_possible_point))*10)
    temp_dict["normalized_p2"].append((1-(df[i].P2.mean()/max_possible_point))*10)
    temp_dict["normalized_p3"].append((1-(df[i].P3.mean()/max_possible_point))*10)
    temp_dict["normalized_p4"].append((1-(df[i].P4.mean()/max_possible_point))*10)
    temp_dict["normalized_p5"].append((1-(df[i].P5.mean()/max_possible_point))*10)
    temp_dict["normalized_p6"].append((1-(df[i].P6.mean()/max_possible_point))*10)
    temp_dict["normalized_p7"].append(None)
    temp_dict["normalized_p8"].append(None)
    
pd.DataFrame(temp_dict)
```

```python
scores = []

for i, year in enumerate(r):
    scores.append((1-(df[i].P1.mean()/max_possible_point))*10)
    scores.append((1-(df[i].P2.mean()/max_possible_point))*10)
    scores.append((1-(df[i].P3.mean()/max_possible_point))*10)
    scores.append((1-(df[i].P4.mean()/max_possible_point))*10)
    scores.append((1-(df[i].P5.mean()/max_possible_point))*10)
    scores.append((1-(df[i].P6.mean()/max_possible_point))*10)
    if year == 2012:
        scores.append((1-(df[i].P7.mean()/max_possible_point))*10)
        scores.append((1-(df[i].P8.mean()/max_possible_point))*10)

print(scores)
```

# <div style="padding:20px;color:white;margin:0;font-size:30px;font-family:Georgia;text-align:left;display:fill;border-radius:5px;background-color:#7b6b59;overflow:hidden">COMBINE</div>


```python
egmo = pd.read_csv("egmo/egmo.csv")
egmo['score'] = scores
```
