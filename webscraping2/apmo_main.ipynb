```python
import numpy as np
import pandas as pd

r = range(2023, 2010-1, -1)
df = pd.read_csv("apmo/apmo_board.csv")
max_possible_point = 7
df
```

```python
# This cell is not used, ONLY for display
temp_dict = {
    "year": list(df.year),
    "normalized_p1": [],
    "normalized_p2": [],
    "normalized_p3": [],
    "normalized_p4": [],
    "normalized_p5": [],
}

for i in range(df.shape[0]):
    temp_dict["normalized_p1"].append((1-(df.p1[i]/max_possible_point))*10)
    temp_dict["normalized_p2"].append((1-(df.p2[i]/max_possible_point))*10)
    temp_dict["normalized_p3"].append((1-(df.p3[i]/max_possible_point))*10)
    temp_dict["normalized_p4"].append((1-(df.p4[i]/max_possible_point))*10)
    temp_dict["normalized_p5"].append((1-(df.p5[i]/max_possible_point))*10)
    
pd.DataFrame(temp_dict)
```

```python
scores = []
for i, year in enumerate(r):
    scores.append((1-(df.p1[i]/max_possible_point))*10)
    scores.append((1-(df.p2[i]/max_possible_point))*10)
    scores.append((1-(df.p3[i]/max_possible_point))*10)
    scores.append((1-(df.p4[i]/max_possible_point))*10)
    scores.append((1-(df.p5[i]/max_possible_point))*10)
    
print(scores)
```

```python
apmo = pd.read_csv('apmo/2023_2010_apmo.csv')
apmo['score'] = scores
```
