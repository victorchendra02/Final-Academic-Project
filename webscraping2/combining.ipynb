```python
# pd.set_option('display.max_rows', 500)

def discretize(value):
    if 0 <= value <= 3:
        return "Easy"
    elif 3 < value <= 6:
        return "Medium"
    elif 6 < value <= 8:
        return "Hard"
    else:
        return "Very Hard"


df_1['score'] = temp
df_1['level'] = df_1['score'].apply(discretize)

df_1.to_csv('imo_combined.csv', index=False)
df_1.sample(6)
```
