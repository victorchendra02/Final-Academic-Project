```python
import pandas as pd


df = pd.read_csv('imo_board.csv')

# Drop NA
df1 = df[["id_key","year","country","all","male","female","p1","p2","p3","p4","p5","p6"]]
df1 = df1.dropna()

# Define aggregation functions
agg_functions = {
    'country': 'count', 
    'all':'sum',
    # 'male':'sum',
    # 'female':'sum',
    'p1': 'sum',
    'p2': 'sum',
    'p3': 'sum',
    'p4': 'sum',
    'p5': 'sum',
    'p6': 'sum',
    # 'p7': 'sum',
    # 'total':'sum'
}

# Group by year and apply aggregation functions
imo_board_groupby_year = df1.groupby('year').agg(agg_functions)

# Rename column 'country'
imo_board_groupby_year.rename(columns={'country': 'count_country'}, inplace=True)

imo_board_groupby_year.to_csv('imo_board_groupby_year_dropna.csv')
display(imo_board_groupby_year)
```
