```python
def decode_html_entities(html_text: str) -> str:
    """
    Decode HTML entities in a given text. Example:
    - `... 4 &amp; 5 ...` --> `... 4 & 5 ...`
    """
    import html

    decoded_text = html.unescape(html_text)
    return decoded_text


def remove_redundant_white_spaces(text: str) -> str:
    from re import sub
    return sub(r'\s+', ' ', text).strip()


def replace_img_tag_with_the_alt_attribute(html_text: str) -> str: 
    """
    Before using this function, must perform:
    - `decode_html_entities()`
    """
    from bs4 import BeautifulSoup
    

    html_text = html_text.replace("\n", " ")
    html_text = html_text.replace("\t", " ")
    # html_text = html_text.replace("\b", " ")
    # html_text = html_text.replace("\v", " ")
    # html_text = html_text.replace("<br>", " ")
    # html_text = html_text.replace("\f", "\\f")
    # html_text = html_text.replace("$$", "$")
    html_text = remove_redundant_white_spaces(html_text)
    
    soup = BeautifulSoup(html_text, 'html.parser')
    img_tags = soup.find_all('img')
    
    # Replace img tags with their alt attribute content
    for img_tag in img_tags:
        img_tag.replace_with(img_tag['alt'])

    modified_text = str(soup)
    # modified_text = modified_text.replace("\a", "\\a")  #
    # modified_text = modified_text.replace("\r", "\\r")
    # modified_text = modified_text.replace("\b", "\\b")  #
    # modified_text = modified_text.replace("&gt;", ">")
    # modified_text = modified_text.replace("&lt;", "<")

    return remove_redundant_white_spaces(modified_text)


def remove_html_tag_in_a_text(html_text: str) -> str:
    from bs4 import BeautifulSoup
    
    clean_text = BeautifulSoup(html_text, 'html.parser').get_text()
    return str(clean_text)


# # Unused
# def list_all_ALT_attribute_values_from_HTML_img_tags(text: str) -> list[str]:
#     from re import findall
#     return findall(r'<img[^>]*alt="([^"]*)"', text)

remove_html_tag_in_a_text('Lorem ipsum, <span>dolor sit</span> amet <i>consectetur</i> <b>adipisicing elit.</b> Quidem, ipsam dolore. <span>Corporis deserunt<br>sint libero, </span>accusantium tempore dolore laborum.')

```

```python
import re


def replace_escape_sequences(text: str):
    def u_escape_sequences(input_string_u):
        # Define a pattern for matching \uXXXX
        pattern = re.compile(r'\\u[0-9a-fA-F]{4}')

        # Replace matched escape sequences with <UNK>
        result_string = re.sub(pattern, '<UNK>', input_string_u)

        return result_string

    def U_escape_sequences(input_string_U):
        # Define a pattern for matching \UXXXXXXXX
        pattern = re.compile(r'\\U[0-9a-fA-F]{8}')

        # Replace matched escape sequences with <UNK>
        result_string = re.sub(pattern, '<UNK>', input_string_U)

        return result_string
    
    
    first = u_escape_sequences(text)
    result = U_escape_sequences(first)
    return result


# Example usage:
# original_string = "<i>Proposed by Gerhard W\\u00f6ginger, Austria.</i>"
# original_string = df[df["year"] == "2022imo_problems"].iloc[-1]["post_rendered"]
# modified_string = replace_escape_sequences(original_string)

# print(modified_string)
```

```python
for i in range(df.shape[0]):
    decoded_text = decode_html_entities(df.iloc[i]['post_rendered'])
    
    extracted_alt_text = replace_img_tag_with_the_alt_attribute(decoded_text)
    final_text = remove_html_tag_in_a_text(extracted_alt_text)
    df.loc[i, ['post_canonical']] = [final_text]
    
```
